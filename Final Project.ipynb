```python
import numpy as np

bh = np.average(df[df['neighbourhood'] == 'Bensonhurst']['longitude'])
bw = np.average(df[df['neighbourhood'] == 'Bushwick']['latitude'])
ch = np.average(df[df['neighbourhood'] == 'Crown Heights']['latitude'])
eh = np.average(df[df['neighbourhood'] == 'East Harlem']['latitude'])
fh = np.average(df[df['neighbourhood'] == 'Fordham']['latitude'])
hl = np.average(df[df['neighbourhood'] == 'Harlem']['latitude'])

df.loc[(df.longitude >= bh) & (df.neighbourhood == 'Bensonhurst'), 'neighbourhood'] = 'Bensonhurst East'
df.loc[(df.longitude < bh) & (df.neighbourhood == 'Bensonhurst'), 'neighbourhood'] = 'Bensonhurst West'

df.loc[(df.latitude >= bw) & (df.neighbourhood == 'Bushwick'), 'neighbourhood'] = 'Bushwick North'
df.loc[(df.latitude < bw) & (df.neighbourhood == 'Bushwick'), 'neighbourhood'] = 'Bushwick South'

df.loc[(df.latitude >= ch) & (df.neighbourhood == 'Crown Heights'), 'neighbourhood'] = 'Crown Heights North'
df.loc[(df.latitude < ch) & (df.neighbourhood == 'Crown Heights'), 'neighbourhood'] = 'Crown Heights South'

df.loc[(df.latitude >= eh) & (df.neighbourhood == 'East Harlem'), 'neighbourhood'] = 'East Harlem North'
df.loc[(df.latitude < eh) & (df.neighbourhood == 'East Harlem'), 'neighbourhood'] = 'East Harlem South'

df.loc[(df.latitude >= fh) & (df.neighbourhood == 'Fordham'), 'neighbourhood'] = 'Bedford Park-Fordham North'
df.loc[(df.latitude < fh) & (df.neighbourhood == 'Fordham'), 'neighbourhood'] = 'Fordham South'

df.loc[(df.latitude >= hl) & (df.neighbourhood == 'Harlem'), 'neighbourhood'] = 'Central Harlem North-Polo Grounds'
df.loc[(df.latitude < hl) & (df.neighbourhood == 'Harlem'), 'neighbourhood'] = 'Central Harlem South'
```

### Introduction

In this project, we are helping a friend plan a hypothetical NYC trip this winter!
This is the first time that our friend is visiting New York, so she wants to explore different Airbnb options during her stay. She's on a slightly tight budget so the cost per night for the place she's staying at is also important too.

First, to have a general idea on how many options are available in New York, we are going to see the distribution based on the 5 big neighborhood groups: Bronx, Brooklyn, Manhattan, Queens, and Staten Island. A higher distribution might mean that people are more willing to stay in that area compared to other places, further suggesting better transportation options and/or close proximity to tourist attractions.

To see if our friend can actually afford these places, we charted a map chart analyzing the average price distribution for each specific neighborhood. This enables us to suggest Airbnbs with a good location, but not too expensive.





### intro
To have an idea of how much our friend should be able to spend on housing, we chose to explore the price range of Airbnbs located in Manhattan. This suggested the ideal budget for her to spend per night, which we decided as $80.


### Description

Manhattan and Brooklyn have the most number of Airbnb units available, with Manhattan having about 200 more. There are not a lot of housing options in the other neighborhoods, so it would be recommended that our friend finds housing in either Manhattan or Brooklyn.



Looking at the Airbnbs in Manhattan, it seems that although there are some very expensive rooms, there are enough affordable rooms priced between $0~100 US Dollars.


### Writeup

During the process of creating this project, we realized that having an end goal or an idea of what we want to gain from the visualization is very important in terms of shaping the whole experience. 

We first had ideas of using a map that shows all the Airbnbs as individual points on a scatterplot map, to make the visualization more intuitive. However, because our dataset was too large (~50k) to efficiently display all the points on the map without overcrowding it, we decided it would be better to show the distribution as a bar chart instead.
